```python
import numpy as np

in_arr = np.array([[1,2,3,0],
                   [0,1,2,3],
                   [3,0,1,2],
                   [2,3,0,1]])
fil_arr = np.array([[2,0,1],
                    [0,1,2],
                    [1,0,2]])
# result = np.zeros((2,2))
out1 = in_arr[:3, :3]
out2 = in_arr[:3,1:4]
out3 = in_arr[1:4,:3]
out4 = in_arr[1:4,1:4]
result1 = np.ravel(fil_arr*out1).sum()
result2 = np.ravel(fil_arr*out2).sum()
result3 = np.ravel(fil_arr*out3).sum()
result4 = np.ravel(fil_arr*out4).sum()

new_arr = np.array([[result1, result2, result3, result4]])
print(new_arr.reshape(2,2))
```
